# Time Series Rolling.

```python
import pandas as pd
```

```python
df = pd.read_csv('Starbucks.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
df['Close'].plot(figsize=(12,5));
```

Add a 7 days **rolling** window.

```python
df.rolling(window=7).mean()
```

```python
df['Close'].plot(figsize=(12,5));
df['Close'].rolling(window=30).mean().plot(figsize=(12,5));
```

```python
df['Close : 30 Day Mean'] = df['Close'].rolling(window=7).mean()
df.head()
```
